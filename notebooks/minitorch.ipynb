```python
import minitorch

from minitorch.module import Module, Parameter
```

```python
class ModuleA1(Module):
    def __init__(self) -> None:
        super().__init__()
        self.p1 = Parameter(5)
        self.non_param = 10
        self.a = ModuleA2()
        self.b = ModuleA3()


class ModuleA2(Module):
    def __init__(self) -> None:
        super().__init__()
        self.p2 = Parameter(10)


class ModuleA3(Module):
    def __init__(self) -> None:
        super().__init__()
        self.c = ModuleA4()


class ModuleA4(Module):
    def __init__(self) -> None:
        super().__init__()
        self.p3 = Parameter(15)
```

```python
mod = ModuleA1()
np = dict(mod.named_parameters())

x = str(mod)
print(x)
assert mod.p1.value == 5
assert mod.non_param == 10

assert np["p1"].value == 5
assert np["a.p2"].value == 10
assert np["b.c.p3"].value == 15
```
